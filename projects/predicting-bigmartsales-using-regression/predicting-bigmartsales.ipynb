```python
import graphlab
```

## Importing the data and dropping NaN values.

```python
lol = graphlab.SFrame('traindata.csv')
train = lol.dropna()
```

```python
train.head()
```

```python
graphlab.canvas.set_target('ipynb')
```

## Checking how parameters in original data are controlling "Item_Output_Sales"

```python
feature = ['Item_Identifier', 'Item_Weight', 'Item_Fat_Content', 'Item_Visibility', 'Item_Type', 'Item_MRP','Outlet_Establishment_Year','Outlet_Size','Outlet_Location_Type','Outlet_Type']
```

```python
train[features].show()
```

```python
train.show(view='BoxWhisker Plot',x='Outlet_Identifier',y='Item_Outlet_Sales')
```

## Building a linear regression model with given features

```python
foo = graphlab.linear_regression.create(train, target='Item_Outlet_Sales', features=feature)
```

```python
test = graphlab.SFrame('testdata.csv')
test.head()
```

```python
results = foo.predict(test)
```
